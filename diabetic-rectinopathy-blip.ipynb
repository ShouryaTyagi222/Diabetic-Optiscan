```python
import os

import requests
from transformers import BlipProcessor, BlipForQuestionAnswering
# from datasets import load_dataset
import torch
from PIL import Image
from torch.utils.data import DataLoader
from tqdm import tqdm
import pickle
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

pretrained_model = 'Salesforce/blip-vqa-base'
model = BlipForQuestionAnswering.from_pretrained(pretrained_model)
processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

torch.cuda.empty_cache()
torch.manual_seed(42)

class VQADataset(torch.utils.data.Dataset):
    def __init__(self, csv_file, img_folder, processor):
        self.data = pd.read_csv(csv_file)
        # self.data = self.data[self.data['image'].str.contains('sharpened')]
        self.img_folder = img_folder
        self.processor = processor
        self.question = "What is the diabetic retinopathy severity in this retinal image?"

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Get image path and label
        image_id = os.path.join(self.img_folder, self.data.iloc[idx, 1])
        answer = self.data.iloc[idx, 2]

        # Load the image
        image = Image.open(image_id).convert("RGB").resize((224,224))
        text = self.question
        encoding = self.processor(image, text, padding="max_length", truncation=True, return_tensors="pt")
        labels = self.processor.tokenizer.encode(
            answer, max_length=12, padding='max_length', return_tensors='pt'
        )
        encoding["labels"] = labels
        for k, v in encoding.items():
            encoding[k] = v.squeeze()

        return encoding
```

```python
dataset = VQADataset(csv_file='/kaggle/working/train_data.csv',
                                     img_folder='',
                                     processor = processor)
batch_size = 4
train_dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, pin_memory=True)

optimizer = torch.optim.AdamW(model.parameters(), lr=4e-5)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.9, last_epoch=-1, verbose=False)
```

```python
num_epochs = 5
model_path = 'blip_model'
patience = 10
min_eval_loss = float("inf")
early_stopping_hook = 0
tracking_information = []
scaler = torch.cuda.amp.GradScaler()
import gc

with open('logs.txt', 'a') as f:
    f.write('Starting the training\n\n')
for epoch in range(num_epochs):
    epoch_loss = 0
    model.train()
    with open('logs.txt', 'a') as f:
        f.write(f'epoch : {epoch}/{num_epochs}\n')
    for idx, batch in zip(tqdm(range(len(train_dataloader)), desc='Training batch: ...'), train_dataloader):
        input_ids = batch.pop('input_ids').to(device)
        pixel_values = batch.pop('pixel_values').to(device)
        attention_masked = batch.pop('attention_mask').to(device)
        labels = batch.pop('labels').to(device)

        with torch.amp.autocast(device_type='cuda', dtype=torch.float16):
            outputs = model(input_ids=input_ids,
                        pixel_values=pixel_values,
                        attention_mask=attention_masked,
                        labels=labels)

        loss = outputs.loss
        epoch_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # scaler.scale(loss).backward()
        # scaler.step(optimizer)
        # scaler.update()
        del input_ids, pixel_values, labels
        gc.collect()
        torch.cuda.empty_cache()

    train_loss = epoch_loss/len(train_dataloader)
    tracking_information.append((train_loss, optimizer.param_groups[0]["lr"]))
    print("Epoch: {} - Training loss: {} - LR: {}".format(epoch+1, train_loss, optimizer.param_groups[0]["lr"]))
    with open('logs.txt', 'a') as f:
        f.write("Epoch: {} - Training loss: {} - LR: {}\n\n".format(epoch+1, train_loss, optimizer.param_groups[0]["lr"]))
    scheduler.step()
    model.save_pretrained(model_path, from_pt=True)
    print("Saved model to", model_path)
    if train_loss < min_eval_loss:
        min_eval_loss = train_loss
        early_stopping_hook = 0
    else:
        early_stopping_hook += 1
        if early_stopping_hook > patience:
            break

    model.eval()
    outputs = model.generate(**inputs)
    answers = [processor.decode(output, skip_special_tokens=True) for output in outputs]
    with open('logs.txt', 'a') as f:
        for i, answer in enumerate(answers):
            f.write(f"Image {i+1}: {answer}\n")
            print(f"Image {i+1}: {answer}")
        f.write('\n')


# pickle.dump(tracking_information, open("tracking_information.pkl", "wb"))
print("The finetuning process has done!")
```

```python
import os
from tqdm import tqdm

data = []

base_path = '/kaggle/input/fundus-images-for-diabetic-retinopathy/OCT diabetic images/train'
for i in os.listdir(base_path):
    folder = base_path + '/' + i
    for j in tqdm(os.listdir(folder)):
#     img = cv2.imread(folder + '/' + j)
#     img = w2d(img, 'rbio1.1',3)
#     cv2.imwrite(new+'/'+j, img)
        data.append([folder + '/' + j,i])

import pandas as pd
df = pd.DataFrame(data, columns=['image', 'label'])
df.to_csv('funds_data.csv')
import os

import requests
from transformers import BlipProcessor, BlipForQuestionAnswering
# from datasets import load_dataset
import torch
from PIL import Image
from torch.utils.data import DataLoader
from tqdm import tqdm
import pickle
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

model = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")
processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

torch.cuda.empty_cache()
torch.manual_seed(42)

class VQADataset(torch.utils.data.Dataset):
    def __init__(self, csv_file, img_folder, processor):
        self.data = pd.read_csv(csv_file)
        self.data = self.data[self.data['image'].str.contains('sharpened')]
        self.img_folder = img_folder
        self.processor = processor
        self.question = "What is the diabetic retinopathy status?"  # Constant question for all images

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Get image path and label
        image_id = os.path.join(self.img_folder, self.data.iloc[idx, 1])
        answer = self.data.iloc[idx, 2]

        # Load the image
        image = Image.open(image_id).convert("RGB").resize((64,64))
        text = self.question
        encoding = self.processor(image, text, padding="max_length", truncation=True, return_tensors="pt")
        labels = self.processor.tokenizer.encode(
            answer, max_length=12, padding='max_length', return_tensors='pt'
        )
        encoding["labels"] = labels
        for k, v in encoding.items():
            encoding[k] = v.squeeze()

        return encoding

# training_dataset = load_dataset("json", data_files="Data/train.jsonl", split="train[:90%]")
# valid_dataset = load_dataset("json", data_files="Data/train.jsonl", split="train[90%:]")
# print("Training sets: {} - Validating set: {}".format(len(training_dataset), len(valid_dataset)))

# train_dataset = VQADataset(dataset=training_dataset,
#                           processor=processor)
# valid_dataset = VQADataset(dataset=valid_dataset,
#                           processor=processor)
dataset = VQADataset(csv_file='/kaggle/working/funds_data.csv',
                                     img_folder='',
                                     processor = processor)
batch_size = 4
train_dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
# train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False, pin_memory=True)
# valid_dataloader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, pin_memory=True)


optimizer = torch.optim.AdamW(model.parameters(), lr=4e-5)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.9, last_epoch=-1, verbose=False)


num_epochs = 1
model_path = 'blip_model'
patience = 10
min_eval_loss = float("inf")
early_stopping_hook = 0
tracking_information = []
scaler = torch.cuda.amp.GradScaler()
import gc

for epoch in range(num_epochs):
    epoch_loss = 0
    model.train()
    for idx, batch in zip(tqdm(range(len(train_dataloader)), desc='Training batch: ...'), train_dataloader):
        input_ids = batch.pop('input_ids').to(device)
        pixel_values = batch.pop('pixel_values').to(device)
        attention_masked = batch.pop('attention_mask').to(device)
        labels = batch.pop('labels').to(device)

        with torch.amp.autocast(device_type='cuda', dtype=torch.float16):
            outputs = model(input_ids=input_ids,
                        pixel_values=pixel_values,
                        attention_mask=attention_masked,
                        labels=labels)

        loss = outputs.loss
        epoch_loss += loss.item()
        # loss.backward()
        # optimizer.step()
        optimizer.zero_grad()

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        del input_ids, pixel_values, labels
        gc.collect()
        torch.cuda.empty_cache()

    train_loss = epoch_loss/len(train_dataloader)
    tracking_information.append((train_loss, optimizer.param_groups[0]["lr"]))
    print("Epoch: {} - Training loss: {} - LR: {}".format(epoch+1, train_loss, optimizer.param_groups[0]["lr"]))
    scheduler.step()
    model.save_pretrained(model_path, from_pt=True)
    print("Saved model to", model_path)
    if train_loss < min_eval_loss:
        min_eval_loss = train_loss
        early_stopping_hook = 0
    else:
        early_stopping_hook += 1
        if early_stopping_hook > patience:
            break

pickle.dump(tracking_information, open("tracking_information.pkl", "wb"))
print("The finetuning process has done!")

image_path = "/kaggle/input/fundus-images-for-diabetic-retinopathy/OCT diabetic images/val/Proliferate/1084_left-rotated2-sharpened.png"
question = "What is the diabetic retinopathy status?"
image = Image.open(image_path)
inputs = processor(images=image, text=question, return_tensors="pt").to(device)
model.eval()
output = model.generate(**inputs)

# Decode the output
answer = processor.decode(output[0], skip_special_tokens=True)
print(answer)


!tar -czvf blip_model.tar.gz /kaggle/working/blip_model


from IPython.display import FileLink
FileLink(r'blip_model.tar.gz')

```
